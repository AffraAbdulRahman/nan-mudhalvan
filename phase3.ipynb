```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd

# Load your CSV file
df=pd.read_csv('C:\\Users\\DELLN5368\\Downloads\\Disease_symptom_and_patient_profile_dataset.csv')


# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Interactive Dashboard "),
    html.Label("Select a column:"),
    dcc.Dropdown(
        id='column-dropdown',
        options=[{'label': col, 'value': col} for col in df.columns],
        value=df.columns[0]
    ),
    html.Div(id='output-container'),
    dcc.Graph(id='graph')
])

# Define callback to update output based on dropdown value
@app.callback(
    Output('output-container', 'children'),
    [Input('column-dropdown', 'value')]
)
def update_output(selected_column):
    return f"You have selected the column: {selected_column}"

# Define callback to update graph based on dropdown value
@app.callback(
    Output('graph', 'figure'),
    [Input('column-dropdown', 'value')]
)
def update_graph(Paitent_ID):
    return {
        'data': [
            {'x': df.index, 'y': df[Paitent_ID], 'type': 'line', 'name': Paitent_ID}
        ],
        'layout': {
            'title': f'{Paitent_ID} over time'
        }
    }

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Assuming the CSV file has a column named 'category' and 'value', change as needed
category_column = 'Gender'
value_column = 'Age'

# Extract values for the pie chart
values = df[value_column]
categories = df[category_column]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart ')

# Displaying the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
