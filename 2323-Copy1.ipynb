```python
import pandas as pd
df=pd.read_csv('C:\\Users\\DELLN5368\\Downloads\\Disease_symptom_and_patient_profile_dataset.csv')
user_profiles = df.groupby('Patient_ID').agg({'Age': 'mean'})
print(user_profiles)
```

```python
df_fillna = df.fillna(df.mean())


df_ffill = df.ffill()

# Replace null values with backward fill method
df_bfill = df.bfill()

print("\nDataFrame after filling null values with mean:")
print(df_fillna)

print("\nDataFrame after forward fill:")
print(df_ffill)

print("\nDataFrame after backward fill:")
print(df_bfill)
#dropna(): Drops rows containing any null values.
#fillna(): Fills null values with a specific value, here we used the mean of each column.
#ffill(): Forward fills null values with the last observed value.
#bfill(): Backward fills null values with the next observed value.
```

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Check for data types
print("\nData types:")
print(df.dtypes)
```

```python
import numpy as np
np.random.seed(10)
data = np.random.normal(loc=0, scale=1, size=100)

# Create box plot
plt.boxplot(data)

# Add labels and title
plt.xlabel('categories')
plt.ylabel('Values')
plt.title('Box Plot Example')
```

```python
np.random.seed(10)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)

# Create violin plot
plt.violinplot([data1, data2])

# Add labels and title
plt.xlabel('categories')
plt.ylabel('Values')
plt.title('Violin Plot Example')

# Show plot
plt.show()
```
